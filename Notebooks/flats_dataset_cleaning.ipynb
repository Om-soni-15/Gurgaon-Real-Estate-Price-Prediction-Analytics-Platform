```python
df['balcony'].isnull().sum()
```

```python
df['balcony']=df['balcony'].replace('No Balcony',0)
df['balcony']=df['balcony'].replace('1 Balcony',1)
df['balcony']=df['balcony'].replace('2 Balconies',2)
df['balcony']=df['balcony'].replace('3 Balconies',3)
df['balcony']=df['balcony'].replace('3+ Balconies',4)
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype(int)
```

```python
df['balcony']
```

```python
df.head()
```

```python
df['society'].value_counts()
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df['price'].value_counts()
```

```python
df[df['price'] == 'Price on Request']
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(price):
    if  type(price)== float:
      return price
    else:
      if 'Lac' in price:
       return round(float(price[0])/100,2)
      else:
        return round(float(price[0]),2)
```

```python
df['price']=df['price'].str.split(' ').apply(treat_price)
```

```python
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft']=df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype(float)
```

```python
df.head()
```

```python
df['areaWithType'].value_counts()
```

```python
df['additionalRoom'].fillna('Not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['additionalRoom'].value_counts()
```

```python
df['floorNum'].isnull().sum()
```

```python
df['floorNum']=df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Lower','0').str.replace('Basement','-1').str.extract(r'(-?\d+)')
```

```python
df[df['floorNum'].isnull()]
```

```python
df.head()
```

```python
df['floorNum'].value_counts()
```

```python
df['facing'].value_counts()
```

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df.info()
```

```python
df.insert(loc=4,column='area',value=round((df['price']*10000000)/df['price_per_sqft']))
```

```python
df.insert(loc=1,column='property_type',value='flat')
```

```python
df.head()
```

```python
df.to_csv('flats_cleaned1.csv',index=False)
```
